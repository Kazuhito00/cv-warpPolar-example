```python
import copy
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

image = cv.imread('image.png')
image = cv.circle(image, (224, 206), 185, (0, 255, 0), 2)

plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.show()
```

```python
lin_polar_img = cv.warpPolar(image, (300, 1000), (224, 206), 185, cv.INTER_CUBIC + cv.WARP_POLAR_LINEAR)
```

```python
plt.imshow(cv.cvtColor(lin_polar_img.transpose(1,0,2)[::-1], cv.COLOR_BGR2RGB))
plt.show()
```

```python
harf_image = copy.deepcopy(image[0:206, 0:556])  # [top : bottom, left : right]

plt.imshow(cv.cvtColor(harf_image, cv.COLOR_BGR2RGB))
plt.show()
```

```python
lin_polar_img2 = cv.warpPolar(harf_image, (300, 1000), (224, 206), 185, cv.INTER_CUBIC + cv.WARP_POLAR_LINEAR + cv.WARP_FILL_OUTLIERS)
```

```python
plt.imshow(cv.cvtColor(lin_polar_img2.transpose(1,0,2)[::-1], cv.COLOR_BGR2RGB))
plt.show()
```

```python
quarter_image = copy.deepcopy(image[0:103, 0:556])  # [top : bottom, left : right]

plt.imshow(cv.cvtColor(quarter_image, cv.COLOR_BGR2RGB))
plt.show()
```

```python
lin_polar_img3 = cv.warpPolar(quarter_image, (300, 1000), (224, 206), 185, cv.INTER_CUBIC + cv.WARP_POLAR_LINEAR + cv.WARP_FILL_OUTLIERS)
```

```python
plt.imshow(cv.cvtColor(lin_polar_img3.transpose(1,0,2)[::-1], cv.COLOR_BGR2RGB))
plt.show()
```
