# インポート

```python
import copy
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# 画像読み込み

```python
%matplotlib inline

image = cv.imread('image/image.jpg')
print(image.shape)

plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.show()
```

# 極座標への変換 / 極座標からの逆変換

```python
# キュービック補間 + 外れ値塗りつぶし + 極座標へリニアマッピング
flags = cv.INTER_CUBIC + cv.WARP_FILL_OUTLIERS + cv.WARP_POLAR_LINEAR
linear_polar_image = cv.warpPolar(image, (300, 1000), (480, 270), 220, flags)  # 引き数：画像, 変換後サイズ、中心座標、半径、変換フラグ
```

```python
# 描画(半時計周りに90度回転)
plt.imshow(cv.cvtColor(linear_polar_image.transpose(1,0,2)[::-1], cv.COLOR_BGR2RGB))
plt.show()
```

```python
# キュービック補間 + 外れ値塗りつぶし + 対数極座標へリニアマッピング
flags = cv.INTER_CUBIC + cv.WARP_FILL_OUTLIERS + cv.WARP_POLAR_LOG
log_polar_image = cv.warpPolar(image, (300, 1000), (480, 270), 220, flags)  # 引き数：画像, 変換後サイズ、中心座標、半径、変換フラグ
```

```python
# 描画(半時計周りに90度回転)
plt.imshow(cv.cvtColor(log_polar_image.transpose(1,0,2)[::-1], cv.COLOR_BGR2RGB))
plt.show()
```

```python
# 逆変換(リニア)
flags = cv.INTER_CUBIC + cv.WARP_FILL_OUTLIERS + cv.WARP_POLAR_LINEAR + cv.WARP_INVERSE_MAP
linear_polar_inverse_image = cv.warpPolar(linear_polar_image, (960, 540), (480, 270), 220, flags)
```

```python
# 描画
plt.imshow(cv.cvtColor(linear_polar_inverse_image, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# 逆変換(対数)
flags = cv.INTER_CUBIC + cv.WARP_FILL_OUTLIERS + cv.WARP_POLAR_LOG + cv.WARP_INVERSE_MAP
log_polar_inverse_image = cv.warpPolar(log_polar_image, (960, 540), (480, 270), 220, flags)
```

```python
# 描画
plt.imshow(cv.cvtColor(log_polar_inverse_image, cv.COLOR_BGR2RGB))
plt.show()
```
